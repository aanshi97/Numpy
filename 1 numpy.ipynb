```python
from skimage import io
photo = io.imread('j.png')         # photo is object name
type(photo)
```

```python
photo.shape                  # 148 rows, 341 columns,... maybe 3 color channal
```

```python
import matplotlib.pyplot as plt
plt.imshow(photo)
```

```python
plt.imshow(photo[::-1])           #-1 reverse
```

```python
plt.imshow(photo[:,::-1])            # like mirror
```

```python
plt.imshow(photo[50:110, 100:250])       # 50:110= row , 100:250= column
```

```python
plt.imshow(photo[::1, ::1])      # every of the row and every of the column
```

```python
plt.imshow(photo[::6, ::6])
```

```python
#photo
photo_sin = np.sin(photo)        # sign of every element which madeup numpy array of this photo
photo_sin
```

```python
photo_sin.ndim                # 3D array
```

```python
photo_sin.shape          # 341 rows, 3 columns, 148 
```

```python
print(np.sum(photo))           # sum of all elements in photo, sum of the array
print(np.prod(photo))
print(np.mean(photo))
print(np.std(photo))
print(np.var(photo))
print(np.min(photo))
print(np.max(photo))
print(np.argmin(photo))
print(np.argmax(photo))
```
